```python
import numpy as np
```

토큰 불러오기

```python
import pickle
with open('./data/train_token_1124.pickle', 'rb') as fs:
    train_token = pickle.load(fs)
```

Mecab를 이용한 형태소 분석 

```python
from eunjeon import Mecab
mecab = Mecab()

doPos = lambda x: mecab.pos(x)

train_senetence = list(map(' '.join, train_token))
pos_tagged = list(map(doPos, train_senetence))
```

```python
getPos = lambda data: [pos for morph, pos in data]
pos_feature = list(map(getPos, pos_tagged))
```

```python
pos_feature[0]
```

품사 집합 만들기

```python
pos_set = list(set([y for x in pos_feature for y in x]))
pos_set = sorted(pos_set)
```

One-hot encoding

```python
def posOnehot(sentence):
    output = []
    for word in sentence:
        onehot = np.zeros(len(pos_set))
        ind = pos_set.index(word)
        onehot[ind] = 1 
        output.append(onehot)
    return output

pos_onehot = list(map(posOnehot, pos_feature))
```
